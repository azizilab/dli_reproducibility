```python
import pandas as pd
```

```python
t_cell_clusters = [0, 3, 7, 19, 45]
b_cell_clusters = [1, 11, 24, 25, 35, 36, 40]
nk_cell_clusters = [2, 15, 30]
myeloid_clusters = [4, 8, 21, 26, 28, 29]
aml_clusters = [14, 23, 41, 42]
```

```python
cells_df = pd.read_csv('data/vireo_labels_allpools.csv', index_col='Unnamed: 0')
cells_df['pool'] = cells_df.index.str.slice(0, 4)
cells_df['barcode'] += '-1'
cells_df
```

### KMA5, AH2

```python
pool_df = cells_df[cells_df['pool'] == 'KMA5']
pool_df
```

```python
pool_df['donor_id'].value_counts()
```

```python
sample_df = pool_df[pool_df['donor_id'] == 'tumorAH']
myeloid_aml_cells_df = sample_df[sample_df['cluster_number'].isin(
                            myeloid_clusters + aml_clusters + [-1])]
myeloid_aml_cells_df
```

```python
myeloid_aml_cells_barcodes = myeloid_aml_cells_df[['barcode']]
myeloid_aml_cells_barcodes.to_csv('data/AH2/myeloid_aml_cells_barcodes.tsv', 
                                  index=False, header=False)
myeloid_aml_cells_barcodes
```

```python
normal_cells_df = pool_df[pool_df['cluster_number'].\
                    isin(t_cell_clusters + b_cell_clusters + nk_cell_clusters)]
normal_cells_df
```

```python
cluster_counts = normal_cells_df['cluster_number'].value_counts()
cluster_counts
```

```python
normal_cells_df = normal_cells_df[normal_cells_df['cluster_number'].isin(
                    cluster_counts[cluster_counts > 1].index)]

cluster_counts = normal_cells_df['cluster_number'].value_counts()
cluster_counts
```

```python
normal_cells_df = normal_cells_df[['barcode', 'cluster_number']]
normal_cells_df.columns = ['cell', 'group']
normal_cells_df.to_csv('data/AH2/normal_cells_annotations.csv', index=False)
normal_cells_df
```
